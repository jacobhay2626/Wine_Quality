# Confidence Intervals

Calculating upper and lower confidence intervals for fixed acidity at each quality level for red wine. 

```python
df_quality8=df_red[df_red["quality"] == 8]
df_quality7=df_red[df_red["quality"] == 7]
df_quality6=df_red[df_red["quality"] == 6]
df_quality5=df_red[df_red["quality"] == 5]
df_quality4=df_red[df_red["quality"] == 4]
df_quality3=df_red[df_red["quality"] == 3]

df_quality8_fixedacidity = df_quality8['fixed acidity']
df_quality7_fixedacidity = df_quality7['fixed acidity']
df_quality6_fixedacidity = df_quality6['fixed acidity']
df_quality5_fixedacidity = df_quality5['fixed acidity']
df_quality4_fixedacidity = df_quality4['fixed acidity']
df_quality3_fixedacidity = df_quality3['fixed acidity']

mean8 = np.mean(df_quality8_fixedacidity)
mean7 = np.mean(df_quality7_fixedacidity)
mean6 = np.mean(df_quality6_fixedacidity)
mean5 = np.mean(df_quality5_fixedacidity)
mean4 = np.mean(df_quality4_fixedacidity)
mean3 = np.mean(df_quality3_fixedacidity)

N_8 = np.size(df_quality8_fixedacidity)
N_7 = np.size(df_quality7_fixedacidity)
N_6 = np.size(df_quality6_fixedacidity)
N_5 = np.size(df_quality5_fixedacidity)
N_4 = np.size(df_quality4_fixedacidity)
N_3 = np.size(df_quality3_fixedacidity)

sd_8 = np.std(df_quality8_fixedacidity)
sd_7 = np.std(df_quality7_fixedacidity)
sd_6 = np.std(df_quality6_fixedacidity)
sd_5 = np.std(df_quality5_fixedacidity)
sd_4 = np.std(df_quality4_fixedacidity)
sd_3 = np.std(df_quality3_fixedacidity)

SE_8 = sd_8 / np.sqrt(N_8)
SE_7 = sd_7 / np.sqrt(N_7)
SE_6 = sd_6 / np.sqrt(N_6)
SE_5 = sd_5 / np.sqrt(N_5)
SE_4 = sd_4 / np.sqrt(N_4)
SE_3 = sd_3 / np.sqrt(N_3)

print("(Quality = 8)=> mean: %.2f. N: %.2f. Standard Error: %.2f." % (mean8, N_8, SE_8))
print("(Quality = 7)=> mean: %.2f. N: %.2f. Standard Error: %.2f." % (mean7, N_7, SE_7))
print("(Quality = 6)=> mean: %.2f. N: %.2f. Standard Error: %.2f." % (mean6, N_6, SE_6))
print("(Quality = 5)=> mean %.2f. N: %.2f. Standard Error: %.2f." % (mean5, N_5, SE_5))
print("(Quality = 4)=> mean: %.2f. N: %.2f. Standard Error: %.2f." % (mean4, N_4, SE_4))
print("(Quality = 3)=> mean: %.2f. N: %.2f. Standard Error: %.2f." % (mean3, N_3, SE_3))
```

```python
lower_limit_3 = mean3 - 1.96*SE_3
lower_limit_4 = mean4 - 1.96*SE_4
lower_limit_5 = mean5 - 1.96*SE_5
lower_limit_6 = mean6 - 1.96*SE_6
lower_limit_7 = mean7 - 1.96*SE_7
lower_limit_8 = mean8 - 1.96*SE_8

upper_limit_3 = mean3 + 1.96*SE_3
upper_limit_4 = mean4 + 1.96*SE_4
upper_limit_5 = mean5 + 1.96*SE_5
upper_limit_6 = mean6 + 1.96*SE_6
upper_limit_7 = mean7 + 1.96*SE_7
upper_limit_8 = mean8 + 1.96*SE_8

print("(Quality = 8) => lower limit: \t%.2f. mean: \t%.2f. upper limit: \t%.2f" % (lower_limit_8, mean8, upper_limit_8))
print("(Quality = 7) => lower limit: \t%.2f. mean: \t%.2f. upper limit: \t%.2f" % (lower_limit_7, mean7, upper_limit_7))
print("(Quality = 6) => lower limit: \t%.2f. mean: \t%.2f. upper limit: \t%.2f" % (lower_limit_6, mean6, upper_limit_6))
print("(Quality = 5) => lower limit: \t%.2f. mean: \t%.2f. upper limit: \t%.2f" % (lower_limit_5, mean5, upper_limit_5))
print("(Quality = 4) => lower limit: \t%.2f. mean: \t%.2f. upper limit: \t%.2f" % (lower_limit_4, mean4, upper_limit_4))
print("(Quality = 3) => lower limit: \t%.2f. mean: \t%.2f. upper limit: \t%.2f" % (lower_limit_3, mean3, upper_limit_3))
```

# ANOVA

*A method of deducing any statistical difference between the means of three or more independent groups. Works by analysing the levels of variance within the groups.*
*Why is it important? Once you understand how each independent variables mean is different from the others, you understand which of them has a connection to your dependent variable (quality)*
